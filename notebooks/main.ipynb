```python
def recommend_services(citizen_id, df, grouped_df, cluster_service_map, service_id_to_name, top_n=10):
    # Step 1: Fetch the citizen's record
    citizen_row = df[df['citizen_id'] == citizen_id]
    if citizen_row.empty:
        print(f"No data found for citizen_id: {citizen_id}")
        # return []

    # Step 2: Identify group attributes for cluster matching
    group_cols = ['district_id', 'gender', 'caste', 'age_group', 'religion_group']
    citizen_group_values = citizen_row[group_cols].iloc[0].to_dict()

    # Step 3: Match the citizen's group to get the cluster_id
    cluster_match = grouped_df
    for col in group_cols:
        cluster_match = cluster_match[cluster_match[col] == citizen_group_values[col]]

    if cluster_match.empty:
        print(f"No cluster match found for citizen_id: {citizen_id}")
        # return []

    cluster_id = cluster_match['cluster_id'].iloc[0]

    # Step 4: Get all service columns
    service_columns = [col for col in df.columns if col.startswith('service_')]

    # Step 5: Extract services already used by the citizen
    used_service_ids = [
        int(col.replace('service_', ''))
        for col in service_columns
        if citizen_row[col].iloc[0] > 0
    ]
    used_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in used_service_ids]

    print(f"\n🟩 Services already used by Citizen {citizen_id}:")
    for name in used_service_names:
        print(f"  - {name}")

    # Step 6: Get top services for the cluster, excluding used ones
    top_services = cluster_service_map.get(cluster_id, [])
    recommended_service_ids = [sid for sid in top_services if sid not in used_service_ids][:top_n]
    recommended_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in recommended_service_ids]

    print(f"\n🟦 Top {top_n} Recommended Services for Citizen {citizen_id} :")
    for name in recommended_service_names:
        print(f"  - {name}")

```

```python
def recommend_services(citizen_id, df, grouped_df, cluster_service_map, service_id_to_name, service_df, top_n=5):
    # Step 1: Fetch the citizen's record
    citizen_row = df[df['citizen_id'] == citizen_id]
    if citizen_row.empty:
        print(f"No data found for citizen_id: {citizen_id}")
        return

    # Step 2: Identify group attributes for cluster matching
    group_cols = ['gender', 'caste', 'age_group', 'religion_group']
    citizen_group_values = citizen_row[group_cols].iloc[0].to_dict()

    # Step 3: Match the citizen's group to get the cluster_id
    cluster_match = grouped_df
    for col in group_cols:
        cluster_match = cluster_match[cluster_match[col] == citizen_group_values[col]]

    if cluster_match.empty:
        print(f"No cluster match found for citizen_id: {citizen_id}")
        return

    cluster_id = cluster_match['cluster_id'].iloc[0]

    # Step 4: Get all service columns
    service_columns = [col for col in df.columns if col.startswith('service_')]

    # Step 5: Extract services already used by the citizen
    used_service_ids = [
        int(col.replace('service_', ''))
        for col in service_columns
        if citizen_row[col].iloc[0] > 0
    ]
    used_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in used_service_ids]

    print(f"\n🟩 Services already used by Citizen {citizen_id}:")
    for name in used_service_names:
        print(f"  - {name}")

    # Step 6: Prepare recurrence info from service_df
    recurrence_map = dict(zip(service_df['service_id'], service_df['is_recurrent']))

    # Step 7: Get top services for the cluster, applying all filters
    recommended_service_ids = []
    for sid in cluster_service_map.get(cluster_id, []):
        # If already used and not recurrent, skip
        if sid in used_service_ids and not recurrence_map.get(sid, False):
            continue
        # If already recommended, skip
        if sid in recommended_service_ids:
            continue
        # If already used and recurrent, allow (user can use again)
        # If not used, allow
        recommended_service_ids.append(sid)
        if len(recommended_service_ids) >= top_n:
            break

    recommended_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in recommended_service_ids]

    print(f"\n🟦 Top {top_n} Recommended Services for Citizen {citizen_id} :")
    for name in recommended_service_names:
        print(f"  - {name}")
```

```python
def recommend_services_2(citizen_id, df, grouped_df, cluster_service_map, service_id_to_name, service_df, top_n=5):
    # Step 1: Fetch the citizen's record
    citizen_row = df[df['citizen_id'] == citizen_id]
    if citizen_row.empty:
        print(f"No data found for citizen_id: {citizen_id}")
        return

    # Print citizen's age, gender, caste, religion from citizen_master
    citizen_info_row = citizen_master[citizen_master['citizen_id'] == citizen_id]
    if not citizen_info_row.empty:
        age = citizen_info_row['age'].iloc[0]
        gender = citizen_info_row['gender'].iloc[0]
        caste = citizen_info_row['caste'].iloc[0]
        religion = citizen_info_row['religion'].iloc[0]
        print(f"Citizen Info for {citizen_id}:")
        print(f"  Age: {age}")
        print(f"  Gender: {gender}")
        print(f"  Caste: {caste}")
        print(f"  Religion: {religion}")
    else:
        print(f"No demographic info found for citizen_id: {citizen_id}")

    # Step 2: Identify group attributes for cluster matching
    group_cols = ['gender', 'caste', 'age_group', 'religion_group']
    citizen_group_values = citizen_row[group_cols].iloc[0].to_dict()

    # Step 3: Match the citizen's group to get the cluster_id
    cluster_match = grouped_df
    for col in group_cols:
        cluster_match = cluster_match[cluster_match[col] == citizen_group_values[col]]

    if cluster_match.empty:
        print(f"No cluster match found for citizen_id: {citizen_id}")
        return

    cluster_id = cluster_match['cluster_id'].iloc[0]

    # Step 4: Get all service columns
    service_columns = [col for col in df.columns if col.startswith('service_')]

    # Step 5: Extract services already used by the citizen
    used_service_ids = [
        int(col.replace('service_', ''))
        for col in service_columns
        if citizen_row[col].iloc[0] > 0
    ]
    used_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in used_service_ids]

    print(f"\n🟩 Services already used by Citizen {citizen_id}:")
    for name in used_service_names:
        print(f"  - {name}")

    # Step 6: Prepare recurrence info and age range from service_df
    recurrence_map = dict(zip(service_df['service_id'], service_df['is_recurrent']))
    min_age_map = dict(zip(service_df['service_id'], service_df['min_age']))
    max_age_map = dict(zip(service_df['service_id'], service_df['max_age']))

    # Get citizen's age (from original df, not binned)
    if 'age' in citizen_master.columns:
        citizen_age = int(citizen_master[citizen_master['citizen_id'] == citizen_id]['age'].iloc[0])
    else:
        citizen_age = None

    # Step 7: Get top services for the cluster, applying all filters
    recommended_service_ids = []
    for sid in cluster_service_map.get(cluster_id, []):
        # Age filter
        min_age = min_age_map.get(sid, None)
        max_age = max_age_map.get(sid, None)
        if citizen_age is not None:
            if (min_age is not None and citizen_age < min_age) or (max_age is not None and citizen_age > max_age):
                continue
        # If already used and not recurrent, skip
        if sid in used_service_ids and not recurrence_map.get(sid, False):
            continue
        # If already recommended, skip
        if sid in recommended_service_ids:
            continue
        recommended_service_ids.append(sid)
        if len(recommended_service_ids) >= top_n:
            break

    recommended_service_names = [service_id_to_name.get(sid, f"Unknown Service {sid}") for sid in recommended_service_ids]

    print(f"\n🟦 Top {top_n} Recommended Services for Citizen {citizen_id} :")
    for name in recommended_service_names:
        print(f"  - {name}")
```

```python
# Find all services used in a particular district and print their names

def get_services_used_in_district(df, district_id, service_id_to_name):
    # Filter for the district
    district_df = df[df['district_id'] == district_id]
    # Get all service columns
    service_columns = [col for col in df.columns if col.startswith('service_') and df[col].dtype in [int, float]]
    # Find which services are used (sum > 0)
    used_services = [col for col in service_columns if district_df[col].sum() > 0]
    # Extract service IDs
    used_service_ids = [int(col.replace('service_', '')) for col in used_services]
    # Map to service names
    used_service_names = [service_id_to_name.get(sid, f"service_{sid}") for sid in used_service_ids]
    return used_service_ids, used_service_names

# Example usage:
district_id = 1  # replace with your district_id
service_id_to_name = dict(zip(service_master['service_id'], service_master['service_name']))
used_service_ids, used_service_names = get_services_used_in_district(final_df, district_id, service_id_to_name)

print(f"Services used in district {district_id} (IDs): {used_service_ids}")
print(f"Services used in district {district_id} (Names):")
for name in used_service_names:
    print("-", name)
```

```python
def get_top_services_by_unique_citizens_in_district(final_df, service_master, district_df, district_id):
    # Filter for the given district
    district_df_filtered = final_df[final_df['district_id'] == district_id]
    # Get all service columns
    service_columns = [col for col in final_df.columns if col.startswith('service_') and final_df[col].dtype in [int, float]]
    # For each service, count number of unique citizens who used it at least once
    service_citizen_counts = {col: (district_df_filtered[col] > 0).sum() for col in service_columns}
    # Sort by count descending and take top 10
    top_services = sorted(service_citizen_counts.items(), key=lambda x: x[1], reverse=True)[:20]
    # Prepare mapping from service_id to service_name
    service_id_to_name = dict(zip(service_master['service_id'], service_master['service_name']))
    # Prepare mapping from district_id to district_name
    district_id_to_name = dict(zip(district_df['district_id'], district_df['district_name']))
    district_name = district_id_to_name.get(district_id, f"district_{district_id}")
    results = []
    for col, count in top_services:
        sid = int(col.replace('service_', ''))
        name = service_id_to_name.get(sid, f"service_{sid}")
        results.append({'service_id': sid, 'service_name': name, 'unique_citizens': count})
    return district_name, results

# Example usage:
district_id = 1  # replace with your district_id
district_name, top_services = get_top_services_by_unique_citizens_in_district(final_df, service_master, district, district_id)
print(f"Top 20 services in district {district_id} ({district_name}):")
for idx, entry in enumerate(top_services, 1):
    print(f"{idx}. {entry['service_name']} (service_id={entry['service_id']}): {entry['unique_citizens']} citizens")
```

```python
def get_top_services_by_unique_citizens_in_subdivision(final_df, service_master, sub_div_df, sub_div_id):
    """
    Returns the top 10 services (by unique citizens) for a given sub_div_id.
    """
    # Filter for the given subdivision
    sub_div_df_filtered = final_df[final_df['sub_div_id'] == sub_div_id]
    # Get all service columns
    service_columns = [col for col in final_df.columns if col.startswith('service_') and final_df[col].dtype in [int, float]]
    # For each service, count number of unique citizens who used it at least once
    service_citizen_counts = {col: (sub_div_df_filtered[col] > 0).sum() for col in service_columns}
    # Sort by count descending and take top 10
    top_services = sorted(service_citizen_counts.items(), key=lambda x: x[1], reverse=True)[:10]
    # Prepare mapping from service_id to service_name
    service_id_to_name = dict(zip(service_master['service_id'], service_master['service_name']))
    # Prepare mapping from sub_div_id to sub_div_name
    sub_div_id_to_name = dict(zip(sub_div_df['sub_div_id'], sub_div_df['sub_div_name']))
    sub_div_name = sub_div_id_to_name.get(sub_div_id, f"sub_div_{sub_div_id}")
    results = []
    for col, count in top_services:
        sid = int(col.replace('service_', ''))
        name = service_id_to_name.get(sid, f"service_{sid}")
        results.append({'service_id': sid, 'service_name': name, 'unique_citizens': count})
    return sub_div_name, results

# Example usage:
sub_div_id = 80  # replace with your sub_div_id
sub_div_name, top_services = get_top_services_by_unique_citizens_in_subdivision(final_df, service_master, sub_division, sub_div_id)
print(f"Top 10 services in sub_division {sub_div_id} ({sub_div_name}):")
for idx, entry in enumerate(top_services, 1):
    print(f"{idx}. {entry['service_name']} (service_id={entry['service_id']}): {entry['unique_citizens']} citizens")
```
